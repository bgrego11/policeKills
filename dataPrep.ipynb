```python
import pandas as pd
```

```python
pk = pd.read_csv('PoliceKillingsUS.csv')
```

```python
pk.head()
```

```python
tk = pk.groupby('race').count()[['id']]
tk.rename(columns = {'id':'Total Police Killings'})
```

```python
ua = pk[pk.armed == 'unarmed'].groupby('race').count()
```

Demographic overall populations pulled from http://worldpopulationreview.com/countries/united-states-population/

```python


w = 233657078
b = 40241818
```

```python
def pop(r):
    if r == 'B':
        return b
    elif r == 'W':
        return w
    else:
        return 1
    
```

```python
ua['population'] = ua.index.map(pop)
```

```python
def rate(row):
    return float(row['id'])/float(row['population'])
```

```python
ua['Unarmed Police Killings Rate'] = ua.apply(rate,axis=1)
```

```python
ua
```

```python
wb = ua[(ua.index == 'W') | (ua.index == 'B')]
```

```python
wb['Police Killings by Race per 100k Population Members'] = wb['Unarmed Police Killings Rate']*100000000
```
